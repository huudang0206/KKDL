### Đọc tập dữ liệu rược vang trắng

```
import pandas as pd

# cau 1a
wine_white = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", delimiter=";")
df_wine_white = pd.DataFrame(wine_white)
print(wine_white)
```

### Kiễm tra giá trị của datasets

```
#cau 1b
print(wine_white.shape[0]) # co 4898 phan tu
```

```
print(len(set(wine_white.iloc[:,-1]))) # co 7 gia tri nhan khac nhau
```

### Huấn luyện datasets theo nghi thức Hold-out

```
# cau 1c
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_wine_white.drop(columns=['quality']),df_wine_white.iloc[:,-1], test_size = 2/10.0, random_state=100)

print(len(x_test)) # So luong phan tu trong tap test = 980

print(len(x_train)) # So luong phan tu trong tap hoc = 3918
```

### Sử dụng Cây quyết định để huấn luyện mô hình

```
#cau 1d
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = "entropy", random_state=100)
model.fit(x_train, y_train)
```

### Dự đoán nhãn

```
#cau 1e
# Du doan nhan
y_pred = model.predict(x_test)
print(y_pred)
```

### Đánh giá độ chính xác của mô hình

```
from sklearn.metrics import accuracy_score
print("Do chinh xac tap kiem tra ", accuracy_score(y_test,y_pred)*100)
```

### Đánh giá độ chính xác tổng thể bằng ma trận Confusion Matrix

```
from sklearn.metrics import confusion_matrix
import numpy as np
a = np.unique(wine_white.quality)
print("Do chinh xac tong the ")
print(confusion_matrix(y_test, y_pred, a)) 
```

### Dự đoán nhãn cho 10 phần tử đầu

```
# cau 1f
# Du doan nhan cho 10 phan tu dau tien
y_pred10 = model.predict(x_test[0:10])

print(list(y_test[0:10])) # Nhan thuc te cua 10 phan tu dau la: [7, 5, 6, 6, 6, 6, 6, 6, 6, 6]
print(y_pred10) # Nhan du doan cho 10 phan tu dau la: [7 5 5 6 7 5 6 5 6 6]
```

```
print("Do chinh xac tap kiem tra 10 phan tu dau ", accuracy_score(y_test[0:10],y_pred10)*100) # = 60.0
```

```
print("Do chinh xac tong the cho 10 phan tu dau ")
print(confusion_matrix(y_test[0:10], y_pred10, a))
```
